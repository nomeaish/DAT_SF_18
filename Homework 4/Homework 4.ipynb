### 1) Projectile Problem

#### a) Find a good model to fit this data

Lets plot the different models and see which one is a best fit

```python
# Reading the data from the csv
prodata = pd.read_csv('Projectile.csv')
prodata.columns = ['time', 'height']
prodata.head()
```

```python
x = prodata['time'].values
y = prodata['height'].values

X = sm.add_constant(x, prepend=True)
```

```python
# 1) Linear model
results = sm.OLS(y, X).fit()
intercept, slope = results.params
r2 = results.rsquared
print r2

p = figure(title='Linear Model')
p.circle(x,y,color="blue",size=8)

x1 = range(int(min(x)),int(max(x)+1))
# For loop:
y1 = []
for val in x1:
    y1.append(intercept + slope*val)
    
y1 = [intercept + slope*val for val in x1]

p.line(x1,y1,color='red')

show(p)
```

```python
# 2) Quadratic model
X = np.c_[x**2, x, np.ones(len(x))]
results = sm.OLS(y, X).fit()
r2 = results.rsquared
print r2

p = figure(title='Quadric Model')
p.circle(x, y, size=8,color='blue')

xx = np.linspace(-1.25,1.25,100)
p.line(xx, results.predict(np.vander(xx,3)), color='red')
show(p)
```

```python
results.summary()
```

The linear model is not a good fit as it is always increasing or decreasing and the data is first increasing and then decreasing. Note that in the linear model height starts increasing instead of decreasing after the maximum height is reached. Because of that reason this model is not very appropriate. The quadric model seems to fit the data and the reality of the situation the best. For this model R2 = 0.9738 which is pretty
close to 1 so this supports the decision to use this model.

#### Quadratic model , R^2 = 0.9738.

###6) Use a generalized linear model to predict the quality score of the wine.

```python
# Normalizing the data
data_n = pd.DataFrame()
for col in wine_df.columns:
    data_n[col] = wine_df[col].apply(lambda x: float(x)/(10**(len(str(wine_df[col].max()))-1)) )
data_n.head()
```

```python
x = data_n['alcohol'].values
y = data_n['quality']

X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
intercept, slope = results.params
r2 = results.rsquared
```

```python
p = figure(title='alcohol vs quality')
p.circle(x,y,color="blue",size=8)

x1 = np.linspace(min(x),max(x),50)
# For loop:
y1 = []
for val in x1:
    y1.append(intercept + slope*val)
# This generates y1 again, but with one line of code instead
# List Comprehension:
y1 = [intercept + slope*val for val in x1]

p.line(x1,y1,color='red')
show(p)
```

```python
results.summary()
```

```python
from sklearn.linear_model import Ridge, Lasso

x = data_n['alcohol'].values
y = data_n['quality']

X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
intercept, slope = results.params
r2 = results.rsquared

ridge = Ridge(alpha = .2)
ridge.fit(np.vander(x, 6), y)

lasso = Lasso(alpha = 1)
lasso.fit(np.vander(x, 6), y)

p = figure(title='Blue = True; Red = OLS')
p.circle(x, y, size=8,color='blue')

xx = np.linspace(-1.25,1.25,100)
p.line(xx, results.predict(np.vander(xx,6)), color='red')
p.line(xx, ridge.predict(np.vander(xx,6)), color='green')
p.line(xx, lasso.predict(np.vander(xx,6)), color='cyan')

show(p)
```
